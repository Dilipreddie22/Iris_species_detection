```python
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings('ignore')
iris=pd.read_csv(r"C:\Users\HP\Downloads\Iris.csv")
```

```python
iris
```

```python
iris.head()
```

```python
x=iris.iloc[:,1:5].values
x
```

```python
y=iris.iloc[:,5].values
y
```

```python
#label encoding
from sklearn import preprocessing #sklearn is used for preprocessing,implementing machine learning algorithms
le=preprocessing.LabelEncoder()
y=le.fit_transform(y) #used for balancing the dataset
y
```

```python
#splitting
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2)
x_test
x_train
```

```python
# logistic regression
from sklearn.linear_model import LogisticRegression
lr=LogisticRegression()
lr.fit(x_train,y_train)
y_pred=lr.predict(x_test)
y_pred
```

```python
from sklearn.metrics import accuracy_score
ac=accuracy_score(y_test,y_pred)*100
print("LOGISTIC REGRESSION ACCURACY=",ac) 
```
